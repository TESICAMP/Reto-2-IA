```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from geopy import distance

datos = pd.read_csv('datos/coordenadas.csv')  
centros = pd.read_csv("datos/resultados.csv")
```

```python
#for i in range(len(centros)):
#    print(centros.iloc[i,0])
for i in range(len(datos)-1):
    nameCol1 = centros.iloc[i,0] 
    colonia1 = datos[datos["Colonia"]==nameCol1]
    colonia1 = (colonia1.iloc[0,1],colonia1.iloc[0,2])

    nameCol2 = centros.iloc[i+1,0] 
    colonia2 = datos[datos["Colonia"]==nameCol2]
    colonia2 = (colonia2.iloc[0,1],colonia2.iloc[0,2])

    distancia = distance.distance(colonia1,colonia2).km
    print(i,nameCol1,distancia)
```
